```python
plt.figure(figsize=(12,6))    #define width and height
top_categories = apps['Category'].value_counts().head(10)   #how many apps exist in each category that is count
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top 10 App Categories by Count')
plt.ylabel('Number of Apps')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data=apps, x='Rating', bins=30, kde=True, color='skyblue')    #Kernel Density Estimate
plt.title('Distribution of App Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Apps')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=apps, x='Type', palette='pastel')
plt.title('Free vs Paid Apps')
plt.xlabel('Type of App')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# barplot
sns.barplot(
    data=apps,
    x='Category',
    y='Rating',
    estimator=np.mean,
    ci=None,
    palette='viridis'
)
plt.xticks(rotation=90)

plt.title('Average App Rating by Category')
plt.xlabel('App Category')
plt.ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=apps,
    x='Reviews',
    y='Rating',
    hue='Type',  # color by Free/Paid
    alpha=0.6
)

plt.xscale('log')  # Log scale helps with readability (as Reviews range is very wide)
plt.title('App Reviews vs Rating')
plt.xlabel('Number of Reviews (Log Scale)')
plt.ylabel('App Rating')
plt.legend(title='App Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
art_design_apps = apps[apps['Category'] == 'ART_AND_DESIGN']

overall_avg_rating = apps['Rating'].mean()
art_design_avg_rating = art_design_apps['Rating'].mean()

print("Overall Average Rating:", round(overall_avg_rating, 2))
print("Art & Design Apps Average Rating:", round(art_design_avg_rating, 2))
```

```python
selected_categories = ['ART_AND_DESIGN', 'GAME', 'TOOLS', 'EDUCATION']
filtered_apps = apps[apps['Category'].isin(selected_categories)]

plt.figure(figsize=(8, 5))
sns.boxplot(data=filtered_apps, x='Category', y='Rating', palette='Set2')
plt.title('App Ratings Comparison for Selected Categories')
plt.xlabel('App Category')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
apps['Installs'] = apps['Installs'].astype(str)

apps['Installs'] = apps['Installs'].str.replace('[+,]', '', regex=True).astype(int)

free_avg = apps[apps['Type'] == 'Free']['Installs'].mean()
paid_avg = apps[apps['Type'] == 'Paid']['Installs'].mean()

print("Average installs for Free apps:", round(free_avg))
print("Average installs for Paid apps:", round(paid_avg))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

avg_installs = apps.groupby('Type')['Installs'].mean().reset_index()

plt.figure(figsize=(6, 4))
sns.barplot(data=avg_installs, x='Type', y='Installs', palette='pastel')
plt.title('Average Installs: Free vs Paid Apps')
plt.ylabel('Average Installs')
plt.xlabel('App Type')
plt.tight_layout()
plt.show()
```

```python
print("Sentiment Types:", user_reviews['Sentiment'].unique())

sentiment_counts = user_reviews['Sentiment'].value_counts()
print("\nSentiment Counts:\n", sentiment_counts)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=user_reviews, x='Sentiment', palette='Set2')
plt.title('User Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

```python
# Merge on the App column
merged_df = pd.merge(user_reviews, apps, on='App')

merged_df.head()
```

```python
# Create rating buckets
merged_df['Rating_Bucket'] = pd.cut(merged_df['Rating'], bins=[0, 2, 3, 4, 5],
                                    labels=['0-2', '2-3', '3-4', '4-5'])

sentiment_by_rating = merged_df.groupby(['Rating_Bucket', 'Sentiment']).size().reset_index(name='Count')

sentiment_by_rating.head()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=sentiment_by_rating, x='Rating_Bucket', y='Count', hue='Sentiment', palette='muted')
plt.title('User Sentiment by App Rating')
plt.xlabel('App Rating Range')
plt.ylabel('Number of Reviews')
plt.tight_layout()
plt.show()
```

```python
top_apps = apps[
    (apps['Rating'] >= 4.5) &
    (apps['Installs'] >= 1_000_000) &
    (apps['Reviews'].astype(str).str.isnumeric())  # Ensure 'Reviews' is clean
]

top_apps['Reviews'] = pd.to_numeric(top_apps['Reviews'])

top_apps = top_apps.sort_values(by=['Rating', 'Installs', 'Reviews'], ascending=False)

top_apps[['App', 'Category', 'Rating', 'Installs', 'Reviews', 'Type']].head(10)
```
